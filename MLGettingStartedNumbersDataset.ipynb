Read in the sample data set called numbers
Then print an overview of the data.

```python
from sklearn import datasets
# this is a comment
digits = datasets.load_digits()
print(digits.data)
```

Print the target. It gives the ground truth for the digit dataset, that is the number corresponding to each digit image that we are trying to learn:

```python
print(digits.target)
```

import the support vector machine library.

Train the svm classifier.

```python
from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(digits.data[:-1], digits.target[:-1])
```

# New section

Use the classifier to detect the class of a digit.

```python
clf.predict(digits.data[-1:])
```
